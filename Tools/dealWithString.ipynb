# 格式化电话号码为 XXX XXXX XXXX

````python
import re

def format_phone_number(phone_number):
    """格式化电话号码为 XXX XXXX XXXX"""
    # 仅保留数字字符，使用正则去除非数字
    digits = re.sub(r'\D', '', phone_number)  # \D 匹配非数字字符并替换为空

    # 限制最多11位数字
    digits = digits[:11]

    # 格式化号码
    formatted_number = ''
    if len(digits) > 0:
        formatted_number = digits[:3]
    if len(digits) > 3:
        formatted_number += ' ' + digits[3:7]
    if len(digits) > 7:
        formatted_number += ' ' + digits[7:]

    return formatted_number


def main():
    phone_number = ''  # 用于存储数字字符
    all_input = ''     # 用于记录所有输入字符
    extra_characters = 0  # 用于统计在获得11个数字后输入的额外字符

    print("请输入电话号码（输入非数字字符会被忽略，最多可输入16个字符）：")
    while True:
        # 实时读取用户输入的单个字符
        char = input("请输入下一个字符：")

        # 记录所有输入的字符
        all_input += char

        # 检查当前有效数字数量是否已经达到11个
        if len(re.sub(r'\D', '', phone_number)) >= 11:
            # 如果已经达到11个有效数字，允许最多再输入5个字符
            extra_characters += 1
            print(f"你输入的所有字符: {all_input}")
            print("已输入11个有效数字。")

            if extra_characters >= 5:
                print("已达到最多输入字符限制，程序退出。")
                break

        else:
            # 如果输入的是数字字符
            if re.match(r'\d', char):  # \d 匹配数字
                phone_number += char

                # 格式化输出
                formatted = format_phone_number(phone_number)
                print(f"你输入的所有字符: {all_input}")
                print(f"格式化后的号码: {formatted}")
            
            else:
                # 非数字字符记录并显示
                print(f"你输入的所有字符: {all_input}")
                print("非数字字符，忽略。")


if __name__ == "__main__":
    main()
````

````python
import re

def is_valid_china_phone_number(number: str) -> bool:
    """
    检查给定的电话号码是否为中国大陆有效的 11 位手机号码。
    
    参数:
    number (str): 要检查的电话号码字符串。
    
    返回:
    bool: 如果是有效的中国大陆 11 位手机号码，返回 True，否则返回 False。
    """
    # 定义中国大陆手机号码的有效段
    valid_prefixes = (
        '134', '135', '136', '137', '138', '139',   # 移动
        '150', '151', '152', '153', '155', '156', '157', '158', '159',   # 移动
        '170', '178', '180', '181', '182', '183', '184', '185', '186', '187', '188', '189',   # 移动
        '130', '131', '132', '133', '134', '135', '136', '137', '138', '139',   # 联通
        '145', '146', '147', '148', '149', '150', '151', '152', '153', '154', '155', '156', '157', '158', '159',   # 联通
        '170', '175', '176', '177', '178', '180', '181', '182', '183', '184', '185', '186', '187', '188', '189',   # 电信
        '133', '153', '177', '180', '181', '182', '183', '184', '185', '186', '187', '188', '189'   # 电信
    )

    # 11 位数字手机号码的正则表达式
    pattern = re.compile(r'^1[3-9]\d{9}$')
    
    # 仅保留数字字符
    digits = re.sub(r'\D', '', number)
    
    # 检查是否是 11 位数字
    if len(digits) != 11:
        return False
    
    # 检查是否符合基本的中国大陆手机号码规则
    if not pattern.match(digits):
        return False
    
    # 检查号码段
    prefix = digits[:3]
    return prefix in valid_prefixes

# 示例调用
print(is_valid_china_phone_number("13400000000"))  # 输出: True
print(is_valid_china_phone_number("13912345678"))  # 输出: True
print(is_valid_china_phone_number("18912345678"))  # 输出: True
print(is_valid_china_phone_number("11912345678"))  # 输出: False
print(is_valid_china_phone_number("12345678901"))  # 输出: False
print(is_valid_china_phone_number("12345"))        # 输出: False
````
